# parametrization ...

\begin{align}
K &= \lambda^{-1} \exp\left(\frac{-r^2}{2l^2}\right) + \sigma_n^2 I\\
&= \lambda^{-1} \exp\left(\frac{-r^2}{2 \times 10^{a}}\right) + 10^b I 
\end{align}

Log marginal likelihood that we optimize is ...
\begin{align}
- \ln L(a, b) = - 1/ 2 \ln (y^T K^{-1} y) - 1 / 2. \ln |det(K)| - n / 2. \ln (2\pi)
\end{align}
