```python
%matplotlib inline
from __future__ import division 
import matplotlib.pyplot as plt
import numpy as np 
from scipy.stats import invgamma
from scipy.special import gamma
```

## writing our own inv-gamma pdf function

The parametrization of the inverse gamma function on [scipy](http://docs.scipy.org/doc/scipy-0.13.0/reference/generated/scipy.stats.invgamma.html) is weird.     
Scipy people fixed $\beta = 1$.    
The most common parametrization is the one shown in [wikipedia](http://en.wikipedia.org/wiki/Inverse-gamma_distribution).   

$$f(x) = \frac{\beta^\alpha}{\Gamma(\alpha)} x^{(-\alpha-1)} \exp\left(-\beta / x \right) $$


```python
x = np.arange(0.01, 3.0, 0.01)
```

```python
plt.plot(x, invgamma.pdf(x, a=1))
plt.plot(x, invgamma.pdf(x, a=2))
plt.plot(x, invgamma.pdf(x, a=3))
```

```python
def invgamma_pdf(x, alpha, beta):
    """pdf of inv gamma dist. with wiki parametrization
    x = np.array 
    alpha = float, real number > 0 
    beta = float, real number > 0 
    
    :math: 
    f(x) = \frac{\beta^\alpha}{\Gamma(\alpha)} 
            x^{(-\alpha-1)} \exp\left(-\beta / x \right)
    """
    return beta ** alpha / gamma(alpha) * \
               x ** (-alpha - 1.) * np.exp(-beta / x)
```

```python
assert np.allclose(invgamma_pdf(x, alpha=1., beta=1.), 
                     invgamma.pdf(x, a=1.)), \
    "something went wrong with the invgamma_pdf function"
```

```python
invgamma_pdf(x, alpha=2., beta=1.) - invgamma.pdf(x, 2.0)
```
